```python
def afficher_N(N):  
    def decorateur(func):
        def wrapper(*args, **kwargs):
            print(f"début N = {N}")
            func(*args, **kwargs)
            print("fin")
        return wrapper
    return decorateur

#@decorateur
def amis(n):
    print(f"j'ai {n} amis !")

amis = afficher_N(4)(amis)

amis(10)
```
